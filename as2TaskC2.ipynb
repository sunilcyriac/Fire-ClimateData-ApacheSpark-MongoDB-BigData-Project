```python
import pymongo
import json
from bson import json_util
from pymongo import MongoClient
def sendRecord(record):
    client = MongoClient()
    db = client.fit5148_db
    collection = db.climate
    cdata = {}
    # Split each record by "||"
    records = record.split("||")
    for item in records:
        # Decode the string to JSON format
        data = json.loads(item)
        # Embedded modeling
        if data['DataType'] == 'cdata':
            cdata = data
            cdata['Fire'] = []
        elif data['DataType'] == 'fdata':
            cdata['Fire'].append(data)
    # Insert the final document into mongodb        
    collection.insert(cdata)
    client.close()
```

```python
import sys

from pyspark import SparkContext
from pyspark.streaming import StreamingContext

# We add this line to avoid an error : "Cannot run multiple SparkContexts at once". If there is an existing spark context, we will reuse it instead of creating a new context.
sc = SparkContext.getOrCreate()

# If there is no existing spark context, we now create a new context
if (sc is None):
    sc = SparkContext(appName="WordCountApp")
ssc = StreamingContext(sc, 5)

host = "localhost"
port = 9999

lines = ssc.socketTextStream(host, int(port))

# Send every batch to sendRecord method for processing
# This line executs for every batch
lines.foreachRDD(lambda rdd: rdd.foreach(sendRecord))

# Print the result                            
lines.pprint()

ssc.start()
try:
    # you can add timeout in the line below like this : "ssc.awaitTermination(timeout = 60)
    # to terminate the connection after 60 seconds
    # right now there is no termination until you interupt the kernel
    ssc.awaitTermination()
except KeyboardInterrupt:
    ssc.stop()
    sc.stop()

ssc.stop()
sc.stop()
```
