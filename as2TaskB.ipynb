**Justification**<br>

The cell below shows the sample of the data model we have designed for the efficient querying of the two data sets  Climate data-Part 1 and Fire data-Part 1.
The model designed is based on the embedded document model. This gives a one to some relationship where some documents are embedded in one document. Here in our case fire data is documents are embedded to the climate data. 

It helps us to avoid the trouble of joining the data sets or documents to get our required result from a query instead you can directly query and retrieve the result. Hence by using embedding model it gives a better read performance, scalability and requires only fewer queries and updates to retrieve a data or update a data.

# Task B.2 and B.3 
**B.2 Create a new database in MongoDB. The new database will be based on the document model you have designed in Task B1.** <br>
**B.3 Write a python program that will read the data from Fire data-Part1 and Climate data -Part1 and load them to the new database created in Task B2.**



```python
import csv
import re
from pprint import pprint
import pymongo
from pymongo import MongoClient

"""
Reading ClimateData-Part1.csv and FireData-Part1.csv as dictionary 
"""

with open('ClimateData-Part1.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    cData = [{k.strip(): int(v) if not re.search('[a-zA-Z.-]+',v) else float(v) if not re.search('[a-zA-Z*-]+',v) else v.strip() for k, v in record.items()} for record in reader]
    
with open('FireData-Part1.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    fire = [{k.strip(): int(v) if not re.search('[a-zA-Z.-]+',v) else float(v) if not re.search('[a-zA-Z*-]+',v) else v.strip() for k, v in d.items()} for d in reader]

"""
Embedded Modeling base on "Date" attribute leads to a single collection of records which
includes FireData inside of a ClimateData where there is a match.
"""            
for c in cData:
        c['Fire'] = []
        for f in fire:
            if c["Date"] == f["Date"]:
                #del f['Date']
                c['Fire'].append(f)
                
"""
****************************************************************************************"""
"""
 connecting to Mongo Client to populate the FireData and ClimateData records into Database
"""                
client = MongoClient()
client.drop_database('fit5148_db')
db = client.fit5148_db
taskbNewCollection = db.climate
#Inserting records into the collection
taskbNewCollection.insert_many(cData)
taskbNewCollObj = taskbNewCollection.find({},{"_id":0})
for c in taskbNewCollObj: 
    pprint(c)
```

```python
"""
TaskB4 - Query of TaskA.2
Find climate data on 15th December 2017.
"""
taskBsearch = taskbNewCollection.find({'Date' : '2017-12-15'} , {"Fire":0 , "_id":0})
for documents in taskBsearch:
    pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA.3
Find the latitude, longitude and confidence when
the surface temperature (C) was between 65 C and 100 C.
"""
taskBrange = taskbNewCollection.aggregate([    
{      
    "$unwind": "$Fire"    
},      
    {"$match" : {"Fire.Surface Temperature (Celcius)": { "$gte":65 , "$lte":100}}} ,
    {"$sort" :{ "Fire.Surface Temperature (Celcius)":1}}, 
{
"$project" : 
{
    "_id":0, 
    "Fire.Latitude":1, 
    "Fire.Longitude":1 , 
    "Fire.Confidence":1,
    "Fire.Surface Temperature (Celcius)":1
} 
} 
])

for documents in taskBrange:
    pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA.4
Find surface temperature (C), air temperature (C), relative humidity 
and maximum wind speed on 15th and 16th of December 2017.
"""
taskBDateRange = taskbNewCollection.find( 
{"Date" : { "$in":["2017-12-15", "2017-12-16"]}},
{
    "_id":0, 
    "Fire.Surface Temperature (Celcius)":1 , 
    "Air Temperature(Celcius)":1 ,
    "Relative Humidity":1,
    "Max Wind Speed":1  
}  
)
                    
for documents in taskBDateRange:
    pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA.5
Find top 10 records with highest surface temperature (C).
just sorted for verifying
"""
taskBrange = taskbNewCollection.aggregate([    
{      
    "$unwind": "$Fire"    
},      
    {"$match" : {"Fire.Confidence" : { "$gte":80 , "$lte":100}}} ,
    {"$sort" :{ "Fire.Confidence":1}}, 
{
"$project" : 
{
    "_id":0, 
    "Date":1, 
    "Fire.Surface Temperature (Celcius)":1 ,
    "Fire.Datetime" : 1,
    "Fire.Confidence" :1,
    "Air Temperature(Celcius)":1
} 
} 
])

for documents in taskBrange:
    pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA6
Find top 10 records with highest surface temperature (C).

"""
taskBsort = taskbNewCollection.aggregate([    
    
    {"$unwind": "$Fire"},      
    {"$sort" :
        { "Fire.Surface Temperature (Celcius)":-1}
    }, 
    {"$limit" : 10 },
    {"$project" : 
        {
            "_id":0, 
            "Fire":1
        } 
    } 
])

for documents in taskBsort:
    pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA.7
Find the number of fire in each day. You are required to only display total number of fire
and the date in the output.
"""

taskBCount = taskbNewCollection.aggregate([
        { "$match": {"Fire": {"$exists": True}}},
        {"$project": {
            "_id":0,
            "Date": 1,
            "numberOfFires": { "$size": "$Fire" }
         }
      }
   ]
)
for documents in taskBCount:
    if documents['numberOfFires'] == 0:
        documents['numberOfFires'] = "No Fire"
    else:
        pprint(documents)
```

```python
"""
TaskB4 - Query of TaskA.8
Find the average surface temperature (C) for each day. You are required to only display
average surface temperature (C) and the date in the output."""


taskBAverage = taskbNewCollection.aggregate([
    {"$match": {"Fire": {"$exists": True}}},
    {"$project": {
            "_id":0,
            "Date": 1,
            "AverageSurfaceTemperature(Celcius)":{"$avg":"$Fire.Surface Temperature (Celcius)"}
         }
      }
    ])
for documents in taskBAverage:
    if documents['AverageSurfaceTemperature(Celcius)'] != None:
        pprint(documents)
```

## Parallel Programming part

```python
"""
TaskB4 - Query of TaskA.2 with Parallel processing
Find climate data on 15th December 2017.
"""

from multiprocessing import Pool

# This is used to return the hash key values
# Intaking the argument value and using mode with processors we can fetch the hash value
def h_hash(x, n):
    result = x%n
    return result



# Hash partition is defined here
def hashPartition(data, processors):
    dic = {}
        
    for record in data: 
        date = record['Date'].split("-")
        hash = h_hash(int(date[2]), processors)  # gets hash of the record
        
        if hash in dic: 
            dic[hash].append(record) 
        else: 
            dic[hash] = [record] 
    
    return dic

def binarySearch ( data , key ):    
    
    # We assume data is sorted , based on analysing the data in the csv files
    # Uncomment the below line if data is unsorted
    #dataSet.sort();
    
    first = 0
    last = len(data) - 1
    
    found = False
    while first <= last and not found:
        
        mid = int((first + last)/2)
        
        if data[mid]['Date'] == key :
            found = True
            
        elif key < data[mid]['Date']:
            last = mid - 1
            
        elif key > data[mid]['Date']:
            first = mid + 1
            
    return data[mid]

def recordSearch(data, querySearch, processors):

    result = []

    pool = Pool(processes = processors)
    
    # Hash Partition based on Date.
    partitionedData = hashPartition(data, processors) 

    date = querySearch.split("-")    
    
    queryHash = h_hash(int(date[2]), processors)
    
    searchPartition = list(partitionedData[queryHash])
    
    #result = searchPartition
    result = pool.apply(binarySearch, [searchPartition, querySearch])
    
    return result

data = taskbNewCollection.find({},{"_id":0})
climateData = []
for record in data:  
    if record.get('Fire'):
        record.pop('Fire')
    climateData.append(record)
    
data = climateData
querySearch = '2017-12-15'
processors = 3
result = recordSearch(data, querySearch, processors)
pprint(result)
```
